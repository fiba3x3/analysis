## [Statistics provided by FIBA](https://worldtour.fiba3x3.com/2022/stats/world-tour/teams)

GP = Games played  
\* W = wins  
\* L = loses  
\* W% = win percentage  
PTS = points  
PPG = points per game  
WBL = Wins before limit  
S-EFF = Shooting efficiency (PTS / (1PTA + 2PTA + FTA))  
S-VAL = Shooting value (S-EFF * PTS)  
S-VALPG = Shooting value per game  
HGL = Highlights  
HGLPG = Highlights per game  
1PTM = One-point field goals made  
1PTA = One-point field goal attempts  
1PT% = One-point field goal percentage  
2PTM = Two-point field goals made  
2PTA = Two-point field goal attempts  
2PT% = Two-point field goal percentage  
FTM = Free throws made  
FTA = Free throw attempts  
FT% = Free throw percentage  
KAS = Key assists  
DRV = Drives  
DNK = Dunks  
BS = Blocked shots  
BZR = Buzzerbeaters  
REB = Rebounds  
REBPG = Rebounds per game  
OREB = Offensive rebounds  
DREB = Defensive rebounds  
TO = Turnovers  
TOPG = Turnovers per game  
PAPG = Points allowed per game  
TF = Team fouls  
TFPG = Team fouls per game  
TFA = Team fouls against  
TFAPG = Team fouls against per game  
2PTAFGA = Two-point field goal attempts per field goal attempts = (2PTA) / field goal attempts (1PTA+2PTA)

## Additional statistics  

1. per game and rebouding  
    POSPG = Possessions per game ([definition of possession](https://nbviewer.org/github/fiba3x3/analysis/blob/main/01_possessions.ipynb) used here differs from FIBA)  
    OREB% = offensive rebound percentage  
    WBLPG = wins before limit per game  

2. key assists  
    KASTO = key assist to turnover ratio  
    KASFGM = key assist to made field goals ratio  

3. points   
    1PTMPTS = percentage of points from 1 point shots  
    2PTMPTS = percentage of points from 2 point shots  
    FTMPTS = percentage of points from free throws  
    DRV1PTM = percentage of made 1 point shots from drives  
    eFG = effective field goal percentage (points per field goal attempt) = (1PTM + 2*2PTM) / (1PTA + 2PTA)  

4. per possession  
    TOPOS = turnovers per possession  
    1PTAPOS = 1 point attempts per possession  
    2PTAPOS = 2 point attempts per possession  
    FGAPOS = field goal attempts (FGA = 1PTA + 2PTA) per possession  
    PPP = points per possession  
    TFAPOS = total fouls against per possession  
    FTAPOS = free throw attempts per possession  
    TFPOS = fouls per possession  
    FTTPOS = free throw trips per possession (percentage of possessions that ended in a trip to free throw)

5. fouls  
    TOTF = turnovers to fouls ratio  
    BSTF = blocked shots to fouls ratio  
    TOTFA = turnovers to fouls against ratio  
    TFTFA = fouls to fouls against ratio  
    FTATFA = free throw attempts to fouls against ratio  
    FTESTFA = extra shot free throws to fouls against ratio  
    FTMTFA = made free throws to fouls against ratio = points per foul  
    FTTTFA = free throw trips to fouls against ratio = (FTA-FTES) / TFA   
    FTESFTA = extra shot free throws to free throws attempt ratio  

\* indiciates this statistic is not available for 2019 and earlier seasons on FIBA's website (without digging through every game played by each team)

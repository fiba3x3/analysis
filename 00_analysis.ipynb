### Piet Van Hasselt Schick Cup 2020 offense outcome by possession

checkball 0.5-0.6 PPP from Schick cup 2020 and analysis of 2019 World cup men's

```python
df_schickcup20 = pd.DataFrame(columns=['1PT', '1PTA', '2PT', '2PTA', 'FT', 'FTA'], index = ['transition','checkball','oreb'],
                            data = [[322,553, 246, 762, 76, 118],[121,191,52,203,12,21],[85,146,19,59,15,19]])
df_schickcup20
```

```python
offense = 'transition'
PPP_tran = (df_schickcup20.loc[offense,'1PT'] + 2*df_schickcup20.loc[offense,'2PT'])/(df_schickcup20.loc[offense,'1PTA'] + df_schickcup20.loc[offense,'2PTA'])
offense = 'checkball'
PPP_check = (df_schickcup20.loc[offense,'1PT'] + 2*df_schickcup20.loc[offense,'2PT'])/(df_schickcup20.loc[offense,'1PTA'] + df_schickcup20.loc[offense,'2PTA'])
offense = 'oreb'
PPP_oreb = (df_schickcup20.loc[offense,'1PT'] + 2*df_schickcup20.loc[offense,'2PT'])/(df_schickcup20.loc[offense,'1PTA'] + df_schickcup20.loc[offense,'2PTA'])
PPP_tran_oreb = (df_schickcup20.loc['transition','1PT'] + df_schickcup20.loc['oreb','1PT'] + 2* df_schickcup20.loc['transition','2PT']+ 2*df_schickcup20.loc['oreb','2PT'])/(df_schickcup20.loc['transition','1PTA'] + df_schickcup20.loc['transition','2PTA'] + df_schickcup20.loc['oreb','1PTA'] + df_schickcup20.loc['oreb','2PTA'])

PPP_tran, PPP_check, PPP_oreb, PPP_tran_oreb
```
